# Customer Segmentation Analysis

## 👥 Overview
Segmentasi pelanggan berdasarkan perilaku belanja.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 📁 Dataset

```
np.random.seed(42)
data = {
    'CustomerID': range(1, 21),
    'Age': np.random.randint(18, 60, 20),
    'SpendingScore': np.random.randint(1, 100, 20),
    'AnnualIncome': np.random.randint(2000, 10000, 20)
}

df = pd.DataFrame(data)
df.head()
```

## 🧠 Segmentasi

```
df['Segment'] = pd.cut(df['SpendingScore'], bins=[0,40,70,100], labels=['Low','Medium','High'])
df.head()
```

## 📊 Visualisasi

```
plt.scatter(df['AnnualIncome'], df['SpendingScore'])
plt.title('Customer Segmentation')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.show()
```

## 💡 Insight
Segmentasi membantu memahami tipe pelanggan untuk strategi marketing.
